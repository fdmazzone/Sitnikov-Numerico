<h3> Resolvente de la EDO </h3>

```julia
λ=1.3
e=.1
m₂=4*(1+λ^2)^(3/2)*(8*λ^3-(1+λ^2)^(3/2))/(64*λ^3-(1+λ^2)^3)
m₁=4*λ^3*(1+λ^2)^(3/2)*(8-(1+λ^2)^(3/2))/(64*λ^3-(1+λ^2)^3)
m=[m₁,m₂]
s=[λ 1]
r,T=Func_r(e)
t₀=0.0
function Fuerza3c(du,u,p,t)
    m=p[1]
    s=p[2]
    r=p[3]
    R = (r(t)^2*s.^2 +u[1]^2).^(1.5)
    du[1] = u[2]
    du[2] = -2*sum(m./R*u[1])
end
alg=Vern9()

function SolveSit(u0,tspan)
    prob_ode_trescuerpos = ODEProblem(Fuerza3c,u0,tspan,[m,s,r]);
    sol = solve(prob_ode_trescuerpos,alg,dt=1e-3,abstol = 1e-12, reltol = 1e-12);
    return sol
end
```

<h2> Fórmula periodo del problema con $e=0$ (al final no lo usé) </h2>

```julia
function PerImpert(z₀)
    rr = (s.^2 +z₀^2).^(.5)
    E=-2*sum(m./rr)
    G(z)=1/sqrt(E+2sum(m./sqrt.(s.^2+z^2)))
    T₀=4*quadgk(G,0,z₀)[1]
    return T₀
end
PerImpert(2.24)/T 
```

<h2> FFT análisis en resonancia 1:1, punto $u_0=(2.466,0)$  </h2>
Vamos a calcular la transformada discreta de fourier en el intervalo $[0,400T]$, en genral a veces en otros intervalos. Vamos a generar una muestra de $N=20\times 400+1$ puntos en ese intervalo.  Por ende $N$  será el orden de la serie tringonomérica involucrada. Los coeficientes del desarrollo se calcula con la transformada rápida y=fft(z), donde $y[n]=\alpha_{n-1}-\beta_{n-1}i$ y   
$$z(t)\sim\frac{1}{N}\left[ \alpha_0+2\sum_{n=1}^{4000}\alpha_{n}\cos\left(\frac{2\pi nt_j}{400T}\right)+\beta_n\sin\left(\frac{2\pi nt_j}{400T}\right)   \right]=
\frac{1}{N}\left[ \alpha_0+2\sum_{n=1}^{4000}\alpha_{n}\cos\left(\frac{2\pi \nu_nt_j}{T}\right)+\beta_n\sin\left(\frac{2\pi \nu_nt_j}{T}\right)   \right],$$

donde $\nu_n=\frac{n}{400}$. Estas frecuencias estan normalizadas para que $\nu=1$ corresponda a periodo $T$. Hemos dividido el espacio de frecuencias en saltos de 1/400=0.0025 y van de 0 a 10. Podemos captar componentes periódicas desde periodos de $400T$ a $T/10$. 



```julia
for i ∈ 1:201; scatter(Z[:,i],V[:,i],s=1);end
xlim(1.75,3)
ylim(-2.5,2.5)
```

```julia
CantPer=100
T_1=CantPer*T
tspan=(0,T_1)
u0=[2.466,0.0]
sol=SolveSit(u0,tspan)
g(t)=sol(t)[1]
```


Usamos FFT y graficamos el módulo de las trasformada

```julia
pygui(true)
t=linspace(0,T_1,CantPer*20+1);
z=g.(t);
y=fft(z)
Y=abs.(y);
j=0:length(Y)-1

#pygui(false)
plot(j/CantPer,Y)

```

Se ve picos de  frecuencia en 1 y todos sus armónicos 0,2,3,4,... lo que es esperable dado que debería ser una solución periódica de frecuencia $\nu=1$. Aparecen dos picos alrededor de 0.9475 y 1.0525.  
<b> Parece que todos los picos de frecuencias son de la forma $k_1\times 1+k_2\times 1.0525$ con $k_1,k_1\in\mathbb{Z}$. Consecuentemente  hay otro pico en una frecuencia baja de 0.0525. Correponde a un periodo de $T/.0525\sim 19T$. </b>.

<b> Debemos encontrar una explicación del porqué de la forma que tienen los picos de frecuencia </b>

La frecuencia de los primarios es 1, que influye en el movimiento de la partícula y es razonable que tenga componentes con esa frecuencia.  Si estuvieramos en la resonancia 1:1 exacta no aparecerían otros picos de  frecuencias más que 1 y sus armónicos 2,3,4,.... En consecuencia, no debemos estar en la resonancia exacta.   

Veamos la aproximación que produce la serie trigonométrica

```julia
I=sortperm(Y[1:2000],rev=true)
J=I[1:20]
deleteat!(J,2)
f(s)=1/length(y)*(y[1]+2*sum(real.(y[J]).*cos.(2*pi*(J-1)*s/T_1)-2*imag.(y[J]).*sin.(2*pi*(J-1)*s/T_1)  ))
t=0:.1:200*T
plot(t/T,g.(t),t/T,f.(t))
xlim(0,50)
```

El periodo de 19T no se ve. Si suavizamos lo veremos

```julia
Y = smooth(g.(t),2000)
pygui(true)
plot(t/T,Y)
```

Este periodo no está en la solución periodica resonante real, pues en esa solución sólo aparecen frecuencias armónicas de 1. Esta frecuencia aparece porque no estamos en la verdadera solución periódica resonante 1:1, estamos recorriendo una curva con forma de banana (muy chiquita) y la recorreremos con periodo aproximadamente de 19T.  Verifiquemos esto.

```julia
h(t)=sol(t)[2]
t1=0:T:19*T
z=g.(t1)
pygui(false)
v=h.(t1)
scatter(z,v)
for i in 0:19
    text(z[i+1],v[i+1],string(i))
end
```

Las principales frecuencias son 1.1075, 0,1075, 0.8925, 1.325. La frecuencia 1.1075 es la principal, es el movimiento cuasi-periódica de ida y vuelta de la partícula. Nuevamente aparecen combinaciones enteras  de esta frcuencia principal y la frecuencia $\nu=1$, es decir $k_1\times 1+k_2\times 1.1075$, $k_1,k_2\in\mathbb{Z}$. 
<b> Pregunta para investigar: Es esto casualidad? Porqué ocurre? </b> 


Ahora podemos explicar porqué aparecen  líneas de trazos con espacios más grandes. Las   frecuencias $\nu=.1075$ y $\nu=1.1075$ determinan los periodos $\frac{T}{1.1075}$ y $\frac{T}{.1075}$. La posición $z(t)$ de la partícula  tiene términos en su serie trigonométrica con funciones con estos dos periódos.  

Cuándo la partícula está nuevamente  cerca de su posición inicial en el diagrama de fases estroboscópico? 

Ahora aparece otro proceso  periódico, este es el  correspondiente a los tiempos en que tomamos las fotos para el mapa estroboscópico. Esto ocurre en los multiplos de  $ T $. Luego para responder la pregunta deberíamos encontrar tres enteros $ k_1, k_2, k_3 $ tales que

$$k_1= \frac{k_2}{0.1075} =\frac{k_3}{1.1075}$$

Por supuesto que hay enteros que resuelven de manera exacta estas ecuaciones pero son muy grandes, originando periodos muy largos que no podremos apreciar en los gráficos nuestros. Tratemos de buscar enteros que las satisfagan aproximadamente dentro de un  rango moderados de enteros. 

```julia
I=[ [i,j,k] for i=1:50, j=1:5, k=1:50 ];
F=x->(x[1]-x[2]/.1075)^2+(x[1]-x[3]/1.1075)^2+(x[2]/.1075-x[3]/1.1075)^2
J=[F(x) for x in I];
J=J[:]
H=sortperm(J)
I[H[1:10]]
```

Veamos los tres primeros tuples [28, 3,31] , [37,4,41] y [9,1,10]. Estos no indica que para $t=28T$ $t=37T$ y $t=9T$, $(z(t),z'(t))$ estaría cerca de $(2.24,0)$. Chequeemos esto.

El periodo de $9T$ corresponde al periodo en que los puntos dan una vuelta a la curva invariante en el retrato de fases, quedan cerca pero no tanto, despues de 4 vueltas completas al retrato de fases, es decir 36T,se retrasa casi el equivalente a un salto de $T$ más.
