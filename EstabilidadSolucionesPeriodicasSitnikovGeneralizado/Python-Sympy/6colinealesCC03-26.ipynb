<h3> Problema de hallar las masas para que la configuracion sea central<h3>


$m_3-----m_2-----m_1-----O-----\underset{\alpha\beta}{m_1}-----\underset{\beta}{m_2}-----\underset{1}{m_3},\ 0<\alpha<1\ y\ 0<\beta<1$




```python
#ecuaciones de configuracion central para las masas m1, m2 y m3
eq1=Eq(m2/(r2-r1)**2-m1/4/r1**2-m2/(r2+r1)**2+m3/(r3-r1)**2-m3/(r3+r1)**2,-lam*r1)
eq2=Eq(-m1/(r2-r1)**2-m2/4/r2**2-m1/(r2+r1)**2+m3/(r3-r2)**2-m3/(r3+r2)**2,-lam*r2)
eq3=Eq(-m1/(r3-r1)**2-m3/4/r3**2-m1/(r3+r1)**2-m2/(r3-r2)**2-m2/(r3+r2)**2,-lam*r3)
eq1, eq2, eq3
```

<h4> Suponemos que las posiciones son $r_1=\alpha\beta$, $r_2=\beta$ y $r_3=1$, con $0<\alpha<1$ y $0<\beta<1$ <h4>

```python
#asumimos r1=alpha*beta, r2=beta r3=1 y lambda=1, 0<alpha<1, 0<beta<1
eq1=eq1.subs(r3,1).subs(r1,alpha*beta).subs(r2,beta).subs(lam,1)
eq2=eq2.subs(r3,1).subs(r1,alpha*beta).subs(r2,beta).subs(lam,1)
eq3=eq3.subs(r3,1).subs(r1,alpha*beta).subs(r2,beta).subs(lam,1)
eq1, eq2, eq3
```

```python
#resolvemos
sol=solve([eq1,eq2,eq3],[m1,m2,m3])
sol
```

```python
simplify(sol[m3])
```
